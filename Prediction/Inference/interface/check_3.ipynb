```python
import xgboost
from IPython.display import display

import shap

# load JS visualization code to notebook
shap.initjs()

# train XGBoost model
X, y = shap.datasets.california(n_points=1_000)
bst = xgboost.train({"learning_rate": 0.01}, xgboost.DMatrix(X, label=y), 100)

# explain the model's predictions using SHAP values
explainer = shap.TreeExplainer(bst)
explanation = explainer(X)

```

```python
import matplotlib.pyplot as plt
shap.plots.force(explanation[0], show=False)

force_plot_html = shap.plots.force(explanation[0], show=False)

# Convertir le contenu HTML en image
# Note : Cette méthode est un peu indirecte et fonctionne mieux dans les environnements Jupyter.
html_str = f"<head>{force_plot_html.header()}</head><body>{force_plot_html.html()}</body>"

# Sauvegarde temporaire en tant que fichier HTML
with open("temp_shap_plot.html", "w") as f:
    f.write(html_str)

# Utilisation de Selenium ou autre méthode pour convertir HTML en image si nécessaire
# Pour cet exemple, nous partons du principe que vous pouvez afficher l'HTML directement
# dans un notebook.

# Affichage du HTML dans une figure matplotlib
fig, ax = plt.subplots(figsize=(10, 4))
ax.axis('off')  # Masquer les axes

# Afficher le contenu HTML dans le notebook
display(HTML(html_str))

# Sauvegarder la figure matplotlib si nécessaire
plt.savefig("shap_plot.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Création des données
x = np.linspace(0, 10, 100)
y1 = np.sin(x)  # Première courbe
y2 = np.exp(x / 10)  # Deuxième courbe avec une autre échelle

# Création de la figure et de l'axe principal
fig, ax1 = plt.subplots()

# Tracé de la première courbe
ax1.plot(x, y1, 'b-', label='sin(x)')
ax1.set_xlabel('X')
ax1.set_ylabel('sin(x)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Création du second axe partageant le même axe x
ax2 = ax1.twinx()

# Tracé de la deuxième courbe
ax2.plot(x, y2, 'r-', label='exp(x/10)')
ax2.set_ylabel('exp(x/10)', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Ajout des légendes
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Affichage du plot
plt.show()
```
