```python
import pandas as pd

data = pd.read_csv('firepoint.csv')

data = data[(data['year'] < 2022) & (data['year'] >= 2017)]

data.dropna(subset='date_debut', inplace=True)
#data.departement.unique()
```

```python
data['Hour'] = data['date_debut'].apply(lambda x : int(x.split(' ')[1].split(':')[0]))
data['Hour'].hist()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Générer un histogramme où toutes les heures de 0 à 23 sont représentées
hour_bins = list(range(0, 25))  # Les bords des bins vont de 0 à 24 pour inclure 23
hour_counts = data['Hour'].value_counts().reindex(range(24), fill_value=0)

# Tracé avec une taille de figure agrandie
plt.figure(figsize=(12, 6))  # Augmente la taille de la figure (largeur=12, hauteur=6)
plt.bar(hour_counts.index, hour_counts.values, color='blue', alpha=0.7)
plt.xticks(range(24))  # Affiche toutes les heures de 0 à 23
plt.xlabel('Hour')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
